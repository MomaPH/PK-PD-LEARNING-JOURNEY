```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def piecewiseplts(t,C0):
  if C0<0.013:
    condls=[t<2.08, t>=2.08] #the conditions
    fnls=[ lambda t: 6*np.exp(t/3), lambda t: 12] # mapping the functions to the conditions
    P= np.piecewise(t, condls, fnls) #generating the overall piecewise condition
  elif C0>=0.013:
    a=5.7*np.log(77*C0)
    b=6.6*np.log(77*C0) + 2.08
    A = (77*C0)**(-2.2)
    condls=[t<a, (a<=t) & (t<b), t>=b]
    fnls=[lambda t: 6*np.exp(-t/3), lambda t: 6*A*np.exp(t/3), lambda t: 12 ]
    P= np.piecewise(t, condls, fnls)
  plt.figure(figsize=(8,5))
  plt.style.use('seaborn-v0_8-white')
  plt.grid()
  plt.plot(t,P)
  plt.title(f'C={C0}')
  plt.xlabel('t(hours)')
  plt.ylabel('P(t)')
  plt.tight_layout()
  plt.show()
```

```python
C= np.array([0, 0.019, 0.038, 0.075, 0.15, 0.3, 0.6, 1.2])
piecewiseplts(t=np.linspace(0,35, num=150)
  , C0=C[:1])
```
